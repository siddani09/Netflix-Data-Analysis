```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.stats import norm

%matplotlib inline
```

```python
# load in the dataset into a pandas dataframe

df=pd.read_csv('C:/Users/sidda/OneDrive/Desktop/netflix_titles.csv')
df.head()
```

```python
#process NaN data
#fill empty director data by add No Director
df.director.fillna("No Director", inplace=True)

#fill emtpy cast data by add No Cast
df.cast.fillna("No Cast", inplace=True)

#fill empty Country data by add Country Unavaible
df.country.fillna("Country Unavailable", inplace=True)

#drop remain NaN data
df.dropna(inplace=True)
```

```python
#check to confirm Null values
df.isnull().sum()
```

```python
#convert time from string to datetime64
df['date_added'] = pd.to_datetime(df['date_added'])

#Extract month and day name and year from Start time after convert
df['added_month']=df['date_added'].dt.month_name()
df['added_day']=df['date_added'].dt.day_name()
df['added_year']=df['date_added'].dt.year
```
